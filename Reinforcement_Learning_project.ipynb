```python
def clean_and_engineer(df):
 
    for col in df.columns:
        if df[col].dtype == 'float64' or df[col].dtype == 'int64':  
            df[col].fillna(df[col].median(), inplace=True)
        elif df[col].dtype == 'object':  
            df[col].fillna(df[col].mode()[0], inplace=True)

  
    numeric_cols = ['Product Price', 'Total Sales']
    for col in numeric_cols:
        lower = df[col].quantile(0.05)
        upper = df[col].quantile(0.95)
        df[col] = df[col].clip(lower=lower, upper=upper)


    df['Price_Lag_1'] = df['Product Price'].shift(1)
    df['Sales_Lag_1'] = df['Total Sales'].shift(1)
    df['Price_Lag_1'].fillna(df['Product Price'].median(), inplace=True)
    df['Sales_Lag_1'].fillna(df['Total Sales'].median(), inplace=True)

    print("Step 4: Adding rolling averages...")
    df['Price_Rolling_Avg_7'] = df['Product Price'].rolling(window=7, min_periods=1).mean()
    df['Sales_Rolling_Avg_7'] = df['Total Sales'].rolling(window=7, min_periods=1).mean()

    median_price = df['Product Price'].median()
    df['Reward'] = (df['Total Sales'] - df['Predicted Sales']) + (df['Product Price'] - median_price)

    scaler = MinMaxScaler()
    df[['Product Price', 'Total Sales', 'Price_Lag_1', 'Sales_Lag_1', 'Price_Rolling_Avg_7', 'Sales_Rolling_Avg_7']] = scaler.fit_transform(
        df[['Product Price', 'Total Sales', 'Price_Lag_1', 'Sales_Lag_1', 'Price_Rolling_Avg_7', 'Sales_Rolling_Avg_7']]
    )

    return df
```

```python
df_soapnuts_cleaned = clean_and_engineer(df_soapnuts.copy())
df_woolball_cleaned = clean_and_engineer(df_woolball.copy())
```

```python
import numpy as np
import gym
from gym import spaces

class ProductPricingEnv(gym.Env):
    def __init__(self, data):
        super(ProductPricingEnv, self).__init__()
        self.data = data
        self.current_step = 0

        self.action_space = spaces.Box(low=0, high=1, shape=(1,), dtype=np.float32)  
        self.observation_space = spaces.Box(low=0, high=1, shape=(6,), dtype=np.float32)  

    def reset(self, seed=None, options=None):
        self.current_step = 0

        if seed is not None:
            np.random.seed(seed)

        return self._next_observation(), {}

    def _next_observation(self):
        obs = self.data.iloc[self.current_step][['Product Price', 'Total Sales', 'Price_Lag_1', 'Sales_Lag_1', 'Price_Rolling_Avg_7', 'Sales_Rolling_Avg_7']].values
        return obs

    def step(self, action):
        new_price = action[0]  
        current_sales = self.data.iloc[self.current_step]['Total Sales']
        predicted_sales = self.data.iloc[self.current_step]['Predicted Sales']

        reward = (new_price * 10) + (current_sales - predicted_sales)  

        self.current_step += 1
        terminated = self.current_step >= len(self.data) - 1  
        truncated = False  

        return self._next_observation(), reward, terminated, truncated, {}
```
